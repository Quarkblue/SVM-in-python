X-Y Split

```python
x = df.loc[:, df.columns != 'Collection']
type(x)
```

```python
y = df['Collection']
type(y)
```

# Test-Train Split
spliting data into 2 parts as test and train as 20% and 80% respectively to validate our model

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
x_train.head()
```

# Standardizing Data
Standardizing is really important because the Support vector machine can only give correct result when we standardize our data since we are calculating distance and distance depends on the scale of each variable

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler().fit(x_train)
```

```python
x_train_std = sc.transform(x_train)
```

```python
x_test_std = sc.transform(x_test)
```

```python
x_test_std
```

```python
x_test
```

# Training Regression Tree

```python
from sklearn.svm import SVR
svr = SVR(kernel='linear', C = 1200)
```

```python
svr.fit(x_train_std, y_train)
```

# Predicting Values using trained model

```python
y_test_pred = svr.predict(x_test_std)
y_train_pred = svr.predict(x_train_std)
```

```python
y_test_pred
```

# Model Performance

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mean_squared_error(y_test, y_test_pred)
# mean error of our predictions
```

```python
r2_score(y_train, y_train_pred)
```

```python
r2_score(y_test, y_test_pred)
```

Ideally your r2 score values should be b/w 0.5-0.9 for a good model
